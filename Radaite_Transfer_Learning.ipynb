```python
import os
import shutil

def copy_and_order_files(folders_to_copy, destination_folder):
    # Ensure the destination directory exists
    os.makedirs(destination_folder, exist_ok=True)

    # Initialize a counter for file numbering
    file_counter = 1

    # Iterate over each folder to copy files from
    for folder in folders_to_copy:
        # Walk through all files in the current folder
        for root, _, files in os.walk(folder):
            for file in files:
                # Create full path to the source file
                source_file_path = os.path.join(root, file)

                # Generate a new file name with an incremental number
                new_file_name = f"{file_counter}{os.path.splitext(file)[1]}"
                destination_file_path = os.path.join(destination_folder, new_file_name)

                # Copy the file to the destination folder with the new name
                shutil.copy2(source_file_path, destination_file_path)

                print(f"Copied {source_file_path} to {destination_file_path}")

                # Increment the file counter
                file_counter += 1

# Example usage
folders_to_copy = ['/content/drive/MyDrive/Radiate_Fine_Tuning/fog',"/content/drive/MyDrive/Radiate_Fine_Tuning/fog_1","/content/drive/MyDrive/Radiate_Fine_Tuning/fog_2"]  # List of folders to copy files from
destination_folder = '/content/drive/MyDrive/radiate/Fine_tuning/foggy'

copy_and_order_files(folders_to_copy, destination_folder)
```

```python
import os
import shutil
import random

# Define your directories
source_base_dir = '/content/drive/MyDrive/radiate/Fine_tuning'
train_base_dir = '/content/drive/MyDrive/radiate/train'
val_base_dir = '/content/drive/MyDrive/radiate/val'        # Base directory to save validation data

# Create base directories if they don't exist
os.makedirs(train_base_dir, exist_ok=True)
os.makedirs(val_base_dir, exist_ok=True)

# List of the four folders in the source directory
categories = ['snowy', 'foggy', 'rainy', 'Normal']

# Split ratio
train_ratio = 0.9  # 80% for training, 20% for validation

# Iterate over each category folder
for category in categories:
    source_dir = os.path.join(source_base_dir, category)
    train_dir = os.path.join(train_base_dir, category)
    val_dir = os.path.join(val_base_dir, category)

    # Create train and val directories for each category if they don't exist
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(val_dir, exist_ok=True)

    # List all files in the current source directory
    files = os.listdir(source_dir)
    print(files)
    # Shuffle the files for randomness
    random.shuffle(files)

    # Calculate split size
    train_size = int(len(files) * train_ratio)

    # Split the files
    train_files = files[:train_size]
    val_files = files[train_size:]

    # Copy the files to their respective directories
    for file in train_files:
        shutil.copy(os.path.join(source_dir, file), os.path.join(train_dir, file))

    for file in val_files:
        shutil.copy(os.path.join(source_dir, file), os.path.join(val_dir, file))

    print(f"Data split and copy completed for category: {category}")

print("All categories processed successfully!")
```

```python
from ultralytics import YOLO
from PIL import Image
import torchvision.transforms as transforms
import os
import numpy as np
from sklearn.metrics import precision_recall_fscore_support

# Load the trained YOLO model
model = YOLO('/content/drive/MyDrive/Radiate_Fine_Tuning/YOLOv8_Runs/snd_run/runs/classify/train/weights/last.pt')

# Define directories
test_dir = '/content/drive/MyDrive/radiate/Dataset_v0.1/test'
categories = ['rainy', 'snowy', 'Normal', 'foggy']
output_dir = '/content/output_from_pred'  # Directory to save transformed images

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Define a transformation for the images
transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),  # Convert grayscale to RGB
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# Function to get predicted class
def get_predicted_class(result):
    probs = result.probs.cpu().numpy()  # Move to CPU and convert to NumPy array
    top1_index = result.probs.top1
    top1_conf = result.probs.top1conf

    # Get the predicted class and confidence
    predicted_class = result.names[top1_index]
    predicted_confidence = top1_conf.item()
    return predicted_class, predicted_confidence

# Collect ground truth and predicted labels
y_true = []
y_pred = []

# Iterate over each category folder
for category in categories:
    category_dir = os.path.join(test_dir, category)
    output_category_dir = os.path.join(output_dir, category)
    os.makedirs(output_category_dir, exist_ok=True)  # Create category directory in output folder

    for image_name in os.listdir(category_dir):
        if image_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(category_dir, image_name)
            # Open and transform the image
            image = Image.open(image_path).convert('RGB')  # Ensure image is in RGB
            image_tensor = transform(image).unsqueeze(0)  # Add batch dimension

            # Convert tensor back to PIL image
            image_transformed = transforms.ToPILImage()(image_tensor.squeeze(0))  # Remove batch dimension

            # Save the transformed image
            output_image_path = os.path.join(output_category_dir, image_name)
            image_transformed.save(output_image_path)

            # Make a prediction
            results = model.predict(source=image_path)  # Use the path to the image

            # Handle results
            for result in results:
                pred_class, _ = get_predicted_class(result)
                true_class = category  # True class is the folder name
                if pred_class is not None:
                    y_true.append(true_class)
                    y_pred.append(pred_class)
                    print(f"Processed {image_name}: Predicted class - {pred_class} with confidence - {result.probs.top1conf.item():.2f}")
                else:
                    print(f"Processed {image_name}: No valid prediction")

# Compute metrics
precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, labels=categories, average='macro')

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

# ResNet Fine Tuning

```python
import torch
from torch import nn, optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import timm

# Step 1: Load the Pre-Trained Model
model = timm.create_model('resnet50.a1_in1k', pretrained=True)

# Step 2: Freeze Layers (Optional)
# Freeze all layers if you only want to fine-tune the last layer
for param in model.parameters():
    param.requires_grad = False

# Alternatively, you can unfreeze specific layers
# e.g., Unfreeze the last 2 layers of the model
for param in list(model.parameters())[-2:]:
    param.requires_grad = True

# Step 3: Modify the Final Layer
num_classes = 4  # Replace with the number of classes in your dataset
model.fc = nn.Linear(model.fc.in_features, num_classes)

# Step 4: Set Up Data Transforms and Data Loaders
data_transforms = {
    'train': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ]),
    'val': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ]),
}

train_dataset = datasets.ImageFolder(root='/content/drive/MyDrive/radiate/Dataset/train', transform=data_transforms['train'])
val_dataset = datasets.ImageFolder(root='/content/drive/MyDrive/radiate/Dataset/val', transform=data_transforms['val'])

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)

# Step 5: Set Up the Training Process
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=0.001)

# Step 6: Train the Model
num_epochs = 10

for epoch in range(num_epochs):
    model.train()  # Set the model to training mode
    running_loss = 0.0
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}')

    # Validate the model
    model.eval()  # Set the model to evaluation mode
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * inputs.size(0)
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    val_loss = val_loss / len(val_loader.dataset)
    accuracy = 100.0 * correct / total
    print(f'Validation Loss: {val_loss:.4f}, Accuracy: {accuracy:.2f}%')

print('Training complete!')
```

```python
from ultralytics import YOLO
import os
import time
import threading
import subprocess

# Function to monitor GPU usage and save it to a log file
def monitor_gpu(log_file):
    with open(log_file, 'w') as f:
        while True:
            # Run the nvidia-smi command to get GPU usage
            result = subprocess.run(['nvidia-smi', '--query-gpu=memory.used,memory.total', '--format=csv,noheader,nounits'],
                                    stdout=subprocess.PIPE,
                                    text=True)
            # Get the output (e.g., "329, 15360")
            gpu_usage = result.stdout.strip()

            # Log the timestamp and memory usage
            log_entry = f"{time.strftime('%Y-%m-%d %H:%M:%S')} - GPU Memory Used: {gpu_usage} MiB\n"
            print(log_entry.strip())
            f.write(log_entry)
            f.flush()  # Ensure the log is written immediately

            # Sleep for 10 seconds before the next check
            time.sleep(10)

# Log file to store GPU memory usage
gpu_log_file = 'gpu_usage_log.txt'

# Start GPU monitoring in a separate thread
monitor_thread = threading.Thread(target=monitor_gpu, args=(gpu_log_file,))
monitor_thread.start()

# Load the pre-trained YOLO model
model = YOLO('yolov8n-cls.pt')

# Fine-tune the model with your custom dataset
model.train(data='/content/drive/MyDrive/radiate/Dataset_v0.1', epochs=10, device=0)
```
